```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(6, 2, figsize = (30, 50), gridspec_kw = {"height_ratios": [6, 1, 6, 1, 6, 1]})

sns.histplot(ax = axis[0, 0], data = filtered_data, x = "age")
sns.boxplot(ax = axis[1, 0], data = filtered_data, x = "age")

sns.histplot(ax = axis[0, 1], data = filtered_data, x = "campaign")
sns.boxplot(ax = axis[1, 1], data = filtered_data, x = "campaign")

sns.histplot(ax = axis[2, 0], data = filtered_data, x = "previous")
sns.boxplot(ax = axis[3, 0], data = filtered_data, x = "previous")

sns.histplot(ax = axis[2, 1], data = filtered_data, x = "cons.conf.idx")
sns.boxplot(ax = axis[3, 1], data = filtered_data, x = "cons.conf.idx")

sns.histplot(ax = axis[4, 0], data = filtered_data, x = "euribor3m")
sns.boxplot(ax = axis[5, 0], data = filtered_data, x = "euribor3m")

sns.histplot(ax = axis[4, 1], data = filtered_data, x = "nr.employed")
sns.boxplot(ax = axis[5, 1], data = filtered_data, x = "nr.employed")

plt.tight_layout()

plt.show()
```

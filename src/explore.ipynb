```python
filtered_data.drop(["day_of_week", "duration", "pdays", "emp.var.rate", "cons.price.idx", "month", "poutcome"], axis = 1, inplace = True)
filtered_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(2, 4, figsize=(50,15))

sns.histplot(ax = axis[0,0], data = filtered_data, x = "job")
sns.histplot(ax = axis[0,1], data = filtered_data, x = "marital").set_xticks([])
sns.histplot(ax = axis[0,2], data = filtered_data, x = "education").set_xticks([])
sns.histplot(ax = axis[0,3], data = filtered_data, x = "default")
sns.histplot(ax = axis[1,0], data = filtered_data, x = "housing")
sns.histplot(ax = axis[1,1], data = filtered_data, x = "loan")
sns.histplot(ax = axis[1,2], data = filtered_data, x = "contact")
sns.histplot(ax = axis[1,3], data = filtered_data, x = "y")

```

Podemos ver que en la mayoría de gráficos tenemos un apartado de "desconocido". Pero en la gran parte de casos tampoco ocupa un espacio grande de datos.
Vemos también que en muchos casos la gran mayoría de las personas están casadas y también vemos que entre créditos, hipotecas y préstamos lo que más ha contestado la gente que si, es a las hipotecas. Podemos llegar a pensar en que ambos casos están relacionados. 
En el apartado de contacto es muy interesante saber que casi el doble de personas han sido contactadas a través del teléfono móvil. 
Por último vemos que alrededor de un 85% de las personas no tienen un depósito a largo plazo.

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axis = plt.subplots(6, 2, figsize = (30, 50), gridspec_kw = {"height_ratios": [6, 1, 6, 1, 6, 1]})

sns.histplot(ax = axis[0, 0], data = filtered_data, x = "age")
sns.boxplot(ax = axis[1, 0], data = filtered_data, x = "age")

sns.histplot(ax = axis[0, 1], data = filtered_data, x = "campaign")
sns.boxplot(ax = axis[1, 1], data = filtered_data, x = "campaign")

sns.histplot(ax = axis[2, 0], data = filtered_data, x = "previous")
sns.boxplot(ax = axis[3, 0], data = filtered_data, x = "previous")

sns.histplot(ax = axis[2, 1], data = filtered_data, x = "cons.conf.idx")
sns.boxplot(ax = axis[3, 1], data = filtered_data, x = "cons.conf.idx")

sns.histplot(ax = axis[4, 0], data = filtered_data, x = "euribor3m")
sns.boxplot(ax = axis[5, 0], data = filtered_data, x = "euribor3m")

sns.histplot(ax = axis[4, 1], data = filtered_data, x = "nr.employed")
sns.boxplot(ax = axis[5, 1], data = filtered_data, x = "nr.employed")

plt.tight_layout()

plt.show()
```
